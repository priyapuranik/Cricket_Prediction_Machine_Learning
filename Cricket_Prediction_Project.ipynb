```python
# creating column total runs

match_data['total_runs'] = match_data['runs_off_bat'] + match_data['extras']
```

```python
match_data.head(2)
```

```python
score = match_data.groupby(['match_id', 'innings']).sum()['total_runs'].reset_index()
```

```python
score = score[score['innings']==1]
score
```

```python
# another dataset

matchdf = match_info.merge(score[['match_id', 'total_runs']], left_on='id', right_on='match_id')
```

```python
teams = [ 'India',
    'England', 
    'New Zealand', 
    'Australia', 
    'Sri Lanka',
    'South Africa', 
    'Bangladesh', 
    'Pakistan', 
    'Nepal', 
    'Afghanistan',
    'West Indies', 
    'Scotland', 
    'United Arab Emirates', 
    'Oman',
    'Netherlands', 
    'Zimbabwe', 
    'United States of America', 
    'Ireland',
    'Canada', 
    'Namibia',
    'Papua New Guinea', 
    'Hong Kong',
    'Kenya',]
```

```python
matchdf = matchdf[matchdf['team1'].isin(teams)]
matchdf = matchdf[matchdf['team2'].isin(teams)]
```

```python
matchdf = matchdf[matchdf['dl_applied']==0]
```

```python
matchdf = matchdf[['match_id','city', 'winner','total_runs']]
```

```python
odi_df = matchdf.merge(match_data, on='match_id')
```

```python
odi_df = odi_df[odi_df['innings'] ==2]
```

```python
odi_df = odi_df.dropna(subset=['city','winner'])
```

```python
print(odi_df['total_runs_y'].dtype)
```

```python
odi_df['currentscore'] = odi_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
odi_df['runs_left'] = odi_df['total_runs_x'] - odi_df['currentscore'] + 1
```

```python
# Calculate balls bowled so far (cumulative count within each innings)
odi_df['balls_bowled'] = odi_df.groupby(['match_id', 'innings']).cumcount() + 1
```

```python
odi_df['balls_left'] = 300 - odi_df['balls_bowled']
```

```python
odi_df['balls_left'] = odi_df['balls_left'].apply(lambda x: max(x, 0))
```

```python
odi_df.columns
```

```python
odi_df.head(4)
```

```python
odi_df['player_dismissed'] = odi_df['player_dismissed'].fillna('0')
```

```python
odi_df['player_dismissed'] = odi_df['player_dismissed'].apply(lambda x:x if x=="0" else "1")
```

```python
#changing the datatye of the column player_dismissed

odi_df['player_dismissed'] = odi_df['player_dismissed'].astype('int')
odi_df
```

```python
wickets = odi_df.groupby('match_id')['player_dismissed'].cumsum()
odi_df['wickets_left'] = 10 - wickets
odi_df.tail(6)
```

```python
#Current run rate

odi_df['crr'] = (odi_df['currentscore']*6)/(300 - odi_df['balls_left'])
```

```python
# required run rate

odi_df['rrr'] = (odi_df['runs_left']*6)/odi_df['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
odi_df['result'] = odi_df.apply(result, axis=1)
```

```python
odi_df.result.value_counts()
```

```python
odi_df.tail(8)
```

```python
# Feature Selection

final = odi_df[['batting_team','bowling_team','city','runs_left','balls_left',
               'wickets_left','total_runs_x','crr','rrr','result']]
```

```python
final = final.sample(final.shape[0])
final
```

```python
final.dropna(inplace=True)
final = final[(final['runs_left'] >= 0) & (final['balls_left'] >= 0)]
final.sample()
```

```python
final.balls_left.value_counts()
```

```python
final.info()
```

## One Hot Encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
categorical_columns = ['batting_team', 'bowling_team', 'city']
ohe = OneHotEncoder(sparse_output=False, drop='first')
```

```python
ohe_data= ohe.fit_transform(final[categorical_columns])
```

```python
ohe_df = pd.DataFrame(ohe_data, columns=ohe.get_feature_names_out(categorical_columns))
```

```python
final = pd.concat([final.drop(categorical_columns, axis=1), ohe_df], axis=1)
```

```python
final.head()
```

```python
final.columns
```

```python
final.isnull().sum()[final.isnull().sum() > 0]
```

```python
final.dropna(inplace=True)
```

```python
import numpy as np
import pandas as pd

# Assuming your DataFrame is named 'final'

# Check and handle infinity values
final.replace([np.inf, -np.inf], np.nan, inplace=True)

# Define a threshold for very large values based on float32 maximum value
very_large_threshold = np.finfo(np.float32).max

# Replace very large values with NaN
final[final > very_large_threshold] = np.nan

# Drop rows with NaN values
final.dropna(inplace=True)

print("Data cleaned from infinity and very large values.")
```

```python
final.head(8)
```

## Model Building

```python
X = final.drop('result', axis=1)
y = final['result']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = RandomForestClassifier(class_weight='balanced')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
